```python
#Librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Carga de archivos y visualización

```python
inmuebles = pd.read_excel("Inmuebles.xlsx", skiprows=1) #Se elimina la primera fila
inmuebles.shape
inmuebles.columns
```

Filtros

```python
#FILTRO 1: Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” de “Referencia”
inmuebles1 = inmuebles.drop(columns=["Referencia"])
inmuebles1.columns
inmuebles1
```

```python
#Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00
filtro2 = inmuebles1[(inmuebles1["Fecha Alta"] == "2007-04-29") | (inmuebles1["Fecha Alta"] == "2004-01-02")]
filtro2
```

```python
#Todos los registros de “ Tipo: parking, local, industrial y piso”
filtro3 = inmuebles1[(inmuebles1["Tipo"] == "Parking") | (inmuebles1["Tipo"] == "Local") | (inmuebles1["Tipo"] == "Industrial") | (inmuebles1["Tipo"] == "Piso")]
filtro3
```

```python
#Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”
filtro4 = inmuebles1[(inmuebles1["Provincia"] == "Girona") & (inmuebles1["Superficie"] > 100)]
filtro4
```

```python
#Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”
filtro5 = inmuebles1[(inmuebles1["Precio Venta"] > 1000000) | (inmuebles1["Vendedor"] == "Carmen") | (inmuebles1["Vendedor"] == "Joaquin")]
filtro5
```

```python
#Todos los registros que no sean de “Provincia : Barcelona“
filtro6 = inmuebles1[(inmuebles1["Provincia"] != "Barcelona")]
filtro6
```

```python
#Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“
filtro7 = inmuebles1[(inmuebles1["Operación"] != "Alquiler") & ~(inmuebles1["Tipo"] == "Oficina")]
filtro7
```

Pasar a CSV

```python
inmuebles1.to_csv("Entregable2.1.csv")
filtro2.to_csv("Entregable2.2.csv")
filtro3.to_csv("Entregable2.3.csv")
filtro4.to_csv("Entregable2.4.csv")
filtro5.to_csv("Entregable2.5.csv")
filtro6.to_csv("Entregable2.6.csv")
filtro7.to_csv("Entregable2.7.csv")
```
